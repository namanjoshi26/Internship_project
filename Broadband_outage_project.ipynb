```python
x=employee_df1.drop(columns=['outage_duration'])         #CREATING TRAINNG AND TESTING DATASET
y=employee_df1['outage_duration']   
print(x.shape)
print(y.shape)
```

```python
x=np.array(x)            
y=np.array(y)
y=y.reshape(-1,1)       #to make y (5904, 1) so that the comparison is with 2D
print(y.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder #Applying one hot encoding
from sklearn.compose import ColumnTransformer
columnTransformer = ColumnTransformer(transformers = [('Encoder', 
                                        OneHotEncoder(sparse=False), 
                                        [0,1,2,3,4])],
                                      remainder='passthrough') 
x=pd.DataFrame(columnTransformer.fit_transform(x))  
x
```

```python
from sklearn.preprocessing import StandardScaler  # scaling the data before training the model
scaler_x= StandardScaler()        # object for the class
x= scaler_x.fit_transform(x)
```

```python
x.size
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis #applying LDA to reduce the dimensonality 
lda_model = LinearDiscriminantAnalysis()
X_lda = lda_model.fit_transform(x,y.ravel())
```

```python
from sklearn.linear_model import LogisticRegression #splitting training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_lda, y, test_size = 0.25,random_state=1)
```

# Logistic Regression(Preferred model as compared to others)

```python
from sklearn.model_selection import GridSearchCV   #applying grid search for hyper parameter tuning
from sklearn.linear_model import LogisticRegression
grid={"C":np.logspace(-3,3,7), "penalty":["l1","l2"]}
logreg=LogisticRegression()
logreg_cv=GridSearchCV(logreg,grid,cv=10)
logreg_cv.fit(X_train,y_train)

print("tuned hpyerparameters :(best parameters) ",logreg_cv.best_params_)
print("accuracy :",logreg_cv.best_score_)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
lr= LogisticRegression(C=0.1,penalty="l2")#using the best parameter according to multiple testing
lr.fit(X_train, y_train.ravel()) #fitting the training data
y_pred = lr.predict(X_test) #prediction of outage
```

```python
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))  #performance metrics
```

```python
y_pred_train = lr.predict(X_train)  
print(confusion_matrix(y_train, y_pred_train))
print(classification_report(y_train, y_pred_train)) #performance metrics of train set
```

### Decision Tree ###

```python
from sklearn.tree import DecisionTreeClassifier
max_depth = []             #pruning the tree using max_depth
acc_gini = []
acc_entropy = []
```

```python
for i in range(1,30):   #checking accuracy of gini and entropy from 1 to 30
 data = DecisionTreeClassifier(criterion='gini', max_depth=i)
 data.fit(X_train, y_train)  
 y_predict = data.predict(X_test)  
 acc_gini.append(accuracy_score(y_test, y_predict))
 
 data = DecisionTreeClassifier(criterion='entropy', max_depth=i) 
 data.fit(X_train, y_train)      #fitting the model
 y_predict = data.predict(X_test) 
 acc_entropy.append(accuracy_score(y_test, y_predict)) #appending entropy and max_depth 
 
 max_depth.append(i)
```

```python
d = pd.DataFrame({'acc_gini':pd.Series(acc_gini), 
 'acc_entropy':pd.Series(acc_entropy),
 'max_depth':pd.Series(max_depth)})
# visualizing changes in parameters
plt.plot('max_depth','acc_gini', data=d, label='gini')
plt.plot('max_depth','acc_entropy', data=d, label='entropy')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend()
```

```python
# training a DescisionTreeClassifier 
from sklearn.tree import DecisionTreeClassifier 
dtree_model = DecisionTreeClassifier(criterion='gini',max_depth = 6).fit(X_train, y_train) # Training the model on the training dataset 
dtree_predictions = dtree_model.predict(X_test) # performing predictions on the test dataset
```

```python
cm = confusion_matrix(y_test, dtree_predictions) 
print(classification_report(y_test,dtree_predictions)) #performance metrics
```

```python
y_pred_train1=dtree_model.predict(X_train)  
print(confusion_matrix(y_train, y_pred_train1))
print(classification_report(y_train, y_pred_train1)) #performance metrics of train set
```
